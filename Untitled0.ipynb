```python
import torch
import numpy as np
from torchvision import datasets
import torchvision.transforms as transforms
```

```python
# convert data to torch.FloatTensor
transform = transforms.ToTensor()

# load the training and test datasets
train_data = datasets.MNIST(root='~/.pytorch/MNIST_data/', train=True, download=True, transform=transform)
test_data = datasets.MNIST(root='~/.pytorch/MNIST_data/', train=False, download=True, transform=transform)
# Create training and test dataloaders
num_workers = 0
# how many samples per batch to load
batch_size = 20

# prepare data loaders
train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size, num_workers=num_workers)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=batch_size, num_workers=num_workers)
```

```python
# MNIST (hand written digit 0 ~ 9)
import matplotlib.pyplot as plt
%matplotlib inline

# obtain one batch of training images
for images, labels in train_loader:
  images = images.numpy()

  # get one image from the batch
  img = np.squeeze(images[0])

  fig = plt.figure(figsize = (5,5)) 
  ax = fig.add_subplot(111)
  ax.imshow(img, cmap='gray')
  break
```

```python
import torch.nn as nn
import torch.nn.functional as F

#Autoencoder(using cnn)
class ConvAutoEncoder(nn.Module):
  def __init__(self):
    super().__init__()

    #encoder
    self.encoder = nn.Sequential(
        nn.Conv2d(1,32,3,padding=1), #input(1channel), output(3channel), filtersize=3, padding=1
        nn.ReLU(True),
        nn.MaxPool2d(2,2),
        nn.Conv2d(32,16, 3, padding=1), #input(32channel), output(16channel), filtersize=3, padding=1
        nn.ReLU(True),
        nn.MaxPool2d(2,2),
        nn.Conv2d(16,8,3,padding=1), #input(16channel), output(8channel), filtersize=3, padding=1
        nn.ReLU(True),
        nn.MaxPool2d(2,2),

    )
    
    #decoder
    self.decoder = nn.Sequential(
        nn.ConvTranspose2d(8,8,3,stride=2), #upsampling 해주는 convolutional layer
        nn.ReLU(True),
        nn.ConvTranspose2d(8,16,2,stride=2),
        nn.ReLU(True),
        nn.ConvTranspose2d(16, 32,2,stride=2),
        nn.ReLU(True),
        nn.Conv2d(32,1,3,padding=1),
        nn.Sigmoid()

    )

  def forward(self,x):
    feature_vector = self.encoder(x)
    reconstructed_x = self.decoder(feature_vector)
    return feature_vector, reconstructed_x

network = ConvAutoEncoder()
dummy_image = torch.randn(1,1,28,28)

feature_vector, reconstructed_x = network(dummy_image)
print(reconstructed_x.shape)
print(reconstructed_x)
```

```python
network = ConvAutoEncoder().cuda() #create network object using cuda

criterion = nn.MSELoss() #mean squared error

optimizer = torch.optim.Adam(network.parameters(),lr=0.001)
```

```python
#max epoch
max_epoch = 20

noise_factor = 0.5

for epoch in range(1, max_epoch+1):
  train_loss = 0.0 #loss value

  network.train()
  for data in train_loader:
    images, _ = data

    noisy_imgs = images + noise_factor*torch.randn(*images.shape)

    noisy_imgs = np.clip(noisy_imgs, 0.0,1.0)
    noisy_imgs = noisy_imgs.cuda()
    images = images.cuda()

    optimizer.zero_grad()
    feature_vector, outputs = network(noisy_imgs)

    loss = criterion(outputs, images)
    loss.backward()
    optimizer.step()

    train_loss += loss.item()

  print('[{}/{}] loss:{:.4f}'.format(epoch, max_epoch, train_loss))
```

```python
for images,labels in test_loader:
  noisy_imgs = images + noise_factor *torch.randn(*images.shape)
  noisy_imgs = np.clip(noisy_imgs, 0.0,1.0)
  noisy_imgs = noisy_imgs.cuda()
  images = images.cuda()

  feature_vector,outputs = network(noisy_imgs)

  #visualization
  noisy_imgs = noisy_imgs.cpu()
  noisy_imgs = noisy_imgs.numpy()

  outputs = outputs.view(batch_size,1,28,28).cpu()
  outputs = outputs.detach().numpy()

  #plot images
  #making images
  #visualization

  fig, axes = plt.subplots(nrows=2, ncols=10, sharex=True, sharey=True, figsize=(25,4))

#noise deleting code
  for noisy_imgs, row in zip([noisy_imgs, outputs],axes):
    for img, ax in zip(noisy_imgs, row):
      ax.imshow(np.squeeze(img), cmap = 'gray')
      ax.get_xaxis().set_visible(False)
      ax.get_yaxis().set_visible(False)

  break
```
